```python
X = df.drop("benign_0__mal_1", axis=1).values
y = df["benign_0__mal_1"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30, activation="relu"))
model.add(Dense(15, activation="relu"))
model.add(Dense(1, activation="sigmoid")) # Binary Classification

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(X_train, y_train, epochs=600, validation_data=(X_test, y_test))
```

```python
pd.DataFrame(model.history.history).plot(figsize=(12,8));
```

We can see overfitting from this graphic, we will use EarlyStopping to solve the overfitting.

```python
model = Sequential()

model.add(Dense(30, activation="relu"))
model.add(Dense(15, activation="relu"))
model.add(Dense(1, activation="sigmoid")) # Binary Classification

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25)
```

```python
model.fit(X_train, y_train, epochs=600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
pd.DataFrame(model.history.history).plot(figsize=(12,8));
```

```python
model = Sequential()

model.add(Dense(30, activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(15, activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(1, activation="sigmoid")) # Binary Classification

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(X_train, y_train, epochs=600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
pd.DataFrame(model.history.history).plot(figsize=(12,8));
```

```python
y_pred = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
